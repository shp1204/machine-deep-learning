```
import pandas as pd
```

```
# raw data import
kospi = pd.read_csv('/content/drive/My Drive/데이터분석 스터디/0_rawdata/kospi200.csv')
samsung = pd.read_csv('/content/drive/My Drive/데이터분석 스터디/0_rawdata/samsung.csv')
```

```
kospi = kospi.rename(columns={'Unnamed: 0':'Time'})
```

```
df = kospi.loc[:,['Time','now']]
datetime = []
for i in df['Time']:
    time = i[:7]
    datetime.append(time)
datetime = pd.DataFrame(datetime)
df2 = pd.concat([df['now'], datetime], axis=1)
df2 = df2.loc[:,[0,'now']]
#df2['now'] = df2['now'].str.replace(',','')
#df2['now'] = df2['now'].astype(int)

result = df2['now'].groupby(df2[0]).mean()
```



---

```
samsung = samsung.rename(columns={'Unnamed: 0':'Time'})
```

```
df = samsung.loc[:,['Time','close']]
datetime = []
for i in df['Time']:
    time = i[:7]
    datetime.append(time)
datetime = pd.DataFrame(datetime)
df2 = pd.concat([df['close'], datetime], axis=1)
df2 = df2.loc[:,[0,'close']]
df2['close'] = df2['close'].str.replace(',','')
df2['close'] = df2['close'].astype(int)

result2 = df2['close'].groupby(df2[0]).mean()
```

```
final_result = pd.concat([result2, result],axis=1)
final_result = final_result.rename(columns={'close':'samsung', 'now':'kospi'})
final_result = final_result.sort_index(ascending = False)
for i in range(len(final_result['kospi'])):
    final_result['kospi'][i] = round(final_result['kospi'][i],4)
```

```
final_result.to_excel('/content/drive/My Drive/데이터분석 스터디/2_result/1_0426.xlsx', encoding='cp949')
```
