```python
len(data)
```

```python
# phone, real email, notes, opt out 가 거의 다 null이므로 column 제거

del_col = ['Phone','Real Email','Notes','Opt Out']

for col in del_col :
    del data[col]
```

## 2 ) 불필요하다고 생각하는 column 제거

1.max orders for one SKU

2.original state는 state와 중복되므로 제거

3.ClientId 제거

```python
del data['Max Orders for one SKU']
del data['OriginalState']
```

```python
data.isnull().sum()
```

```python
data[data['ClientId'].isnull()]
data = data.drop(29238)
```

## 3 ) 공통으로 nan인 83개 데이터 다른 변수로 저장 : not_yet

```python
not_yet = data[data['# of Orders Placed'].isnull()]
tmp = not_yet.index
data = data.drop(tmp)
```

```python
data.isnull().sum()
```

## 4 ) 배송 지역이 nan인 7개 데이터 다른 변수로 저장 : no_delivery

```python
no_delivery = data[data['City'].isnull()]
tmp = no_delivery.index
data = data.drop(tmp)
```

```python
data.isnull().sum()
```

## 2 ) US의 어떤 STATE ?
위에서 US가 가장 많았음으로 US만 분석하자

```python
data_US = data[data['Country'] == 'US']
```

### VI

```python
data_US.groupby(['State']).mean().sort_values(by = 'Total Value of all Orders Placed', ascending = False)
```

State 별로 평균 주문 단가가 높은 지역은 VI이다.

```python
data_US[data_US['State']=='VI']
```

그냥 한 번 주문금액이 컸을 뿐이었다 ;

### CA

```python
data_US.groupby(['State']).sum().sort_values(by = 'Total Value of all Orders Placed', ascending = False)
```

Statae별로 총 주문 금액이 높은 지역은 CA이다.

```python
data_US_CA = data_US[data_US['State']=='CA']
data_US_CA.groupby(['City']).mean().sort_values(by = 'Total Value of all Orders Placed', ascending = False)
```

```python
data_US_CA.groupby(['City']).sum().sort_values(by = 'Total Value of all Orders Placed', ascending = False)
```
